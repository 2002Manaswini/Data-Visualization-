### Feature Engineering

```python
df["Date_Of_Journey"]=pd.to_datetime(df["Date_of_Journey"])
```

```python
df["Date_Of_Journey"].dtype
```

```python
df["Month"]=df["Date_Of_Journey"].dt.month
df["Year"]=df["Date_Of_Journey"].dt.year
df["Date"]=df["Date_Of_Journey"].dt.day
```

```python
df.info()
```

```python
df.drop('Date_Of_Journey',axis=1,inplace=True)
```

```python
df.columns
```

```python
df.head(5)
```

```python
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df.columns
```

```python
df["Arrival_Time"]=df['Arrival_Time'].str.split(' ').str[0]
```

```python
df["Arrival_Time"]
```

```python
df["Arrival_Hr"]=df["Arrival_Time"].str.split(':').str[0]
df["Arrival_Min"]=df["Arrival_Time"].str.split(':').str[1]
```

```python
df['Arrival_Hr']=df['Arrival_Hr'].astype(int)
df['Arrival_Min']=df['Arrival_Min'].astype(int)
```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df.head(5)
```

```python
df["Dep_Hr"]=df["Dep_Time"].str.split(':').str[0]
df["Dep_Min"]=df["Dep_Time"].str.split(':').str[1]
```

```python
df["Dep_Hr"]=df["Dep_Hr"].astype(int)
df["Dep_Min"]=df["Dep_Min"].astype(int)
```

```python
df.drop('Dep_Time',axis=1,inplace=True)
```

```python
df.head(5)
```

```python
df["Total_Stops"].unique()
```

```python
df["Total_Stops"]=df["Total_Stops"].map({'non-stop':0, '2 stops':2, '1 stop':1, '3 stops':3, 'nan' :1, '4 stops':4})
```

```python
df.head(5)
```

```python
df.drop('Route',axis=1,inplace=True)
```

```python
df.head(5)
```

```python
df['Duration_Hr']=df['Duration'].str.split(' ').str[0].str.split('h').str[0]
```
